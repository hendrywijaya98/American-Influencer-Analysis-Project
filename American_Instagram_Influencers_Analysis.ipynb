## Change Column Data Type

### Change Followers Column from Categorical to Numerical

```python
print(influencer_df['Followers'].head(15), '\n')
# df[conditions]['column name']
print(influencer_df[influencer_df['Followers'].str.contains('million')]['Followers'].head(15), '\n')
print(influencer_df[influencer_df['Followers'].str.contains('thousand')]['Followers'].head(15), '\n')
```

Followers column masih dalam bentuk categorical mau diubah jadi numerical
- misal `xx thousand` diubah jadi `xx000`
- misal `xx million` diubah jadi `xx000000` 

```python
foll_temp = []
# iterate df column
for colname, colval in influencer_df.iterrows():
    # if df follower == str thousands
    if 'thousand' in colval['Followers']:
        # remove "thousands" menggunakan replace 
        num1 = colval['Followers'].replace('thousand','')
        # convert into numeric and multiple by 1000
        num1 = int(float(num1)*1000)
        foll_temp.append(num1)
    # elif df follower == str million
    else:
        # remove "million" menggunakan replace 
        num2 = colval['Followers'].replace('million','')
        # convert into numeric and multiple by 1000000
        num2 = int(float(num2) * 1000000)
        foll_temp.append(num2)
 
foll_temp = pd.Series(foll_temp)
influencer_df['Followers'] = foll_temp
influencer_df.head(15)
```

### Change Datatype Object to Datetime

```python
influencer_df['Account Created'] = pd.to_datetime(influencer_df['Account Created'])
influencer_df['Account Created'].head(15)
```
